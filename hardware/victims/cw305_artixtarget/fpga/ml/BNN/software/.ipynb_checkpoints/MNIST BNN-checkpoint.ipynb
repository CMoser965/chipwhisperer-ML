```python
import tensorflow as tf 
import larq as lq
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# Normalize pixel values to be between -1 and 1
train_images, test_images = train_images / 127.5 - 1, test_images / 127.5 - 1
```

```python
# All quantized layers except the first will use the same options
kwargs = dict(input_quantizer="ste_sign",
              kernel_quantizer="ste_sign",
              kernel_constraint="weight_clip")

model = tf.keras.models.Sequential()

# In the first layer we only quantize the weights and not the input
model.add(lq.layers.QuantConv2D(32, (3, 3),
                                kernel_quantizer="ste_sign",
                                kernel_constraint="weight_clip",
                                use_bias=False,
                                input_shape=(28, 28, 1)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.BatchNormalization(scale=False))

model.add(lq.layers.QuantConv2D(64, (3, 3), use_bias=False, **kwargs))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.BatchNormalization(scale=False))

model.add(lq.layers.QuantConv2D(64, (3, 3), use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(tf.keras.layers.Flatten())

model.add(lq.layers.QuantDense(64, use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(lq.layers.QuantDense(10, use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(tf.keras.layers.Activation("softmax"))
```

```python
lq.models.summary(model)
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, batch_size=64, epochs=6)

test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```python
print(f"Test accuracy {test_acc * 100:.2f} %")
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# Normalize pixel values to be between -1 and 1
train_images, test_images = train_images / 127.5 - 1, test_images / 127.5 - 1

import cv2

tiny_trainer = [cv2.resize(x, dsize=(8, 8), interpolation=cv2.INTER_AREA) for x in train_images]
test_trainer = [cv2.resize(x, dsize=(8, 8), interpolation=cv2.INTER_AREA) for x in test_images]

tiny_trainer = np.reshape(tiny_trainer, (60000, 8, 8, 1))
test_trainer = np.reshape(test_trainer, (10000, 8, 8, 1))
```

```python
# All quantized layers except the first will use the same options
kwargs = dict(input_quantizer="ste_sign",
              kernel_quantizer="ste_sign",
              kernel_constraint="weight_clip")

model = tf.keras.models.Sequential()

# In the first layer we only quantize the weights and not the input
model.add(lq.layers.QuantConv2D(32, (3, 3),
                                kernel_quantizer="ste_sign",
                                kernel_constraint="weight_clip",
                                use_bias=False,
                                input_shape=(8, 8, 1)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(tf.keras.layers.Flatten())

model.add(lq.layers.QuantDense(64, use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(lq.layers.QuantDense(10, use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(tf.keras.layers.Activation("softmax"))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(tiny_trainer, train_labels, batch_size=64, epochs=20)

test_loss, test_acc = model.evaluate(test_trainer, test_labels)
```

```python
model.summary()
layers = {}
for layer in model.layers:
    print(layer.name,"quantized weights:",np.array(layer.weights).shape)
    if "batch_norm" not in layer.name:
        layers[layer.name] = np.digitize(layer.weights, bins=[0])
    else:
        layers[layer.name] = layer.weights
```

```python
for layer in layers:
    if "max_pool" in layer or "activation" in layer or "flatten" in layer:
        continue
    print("---------------------<>\t" + str(layer) + "\t<>---------------------")
    if 'quant_conv2d' in layer:
        print(np.shape(layers[layer]))
        print('{', end="")
        for i, weight_vec in enumerate(np.reshape(layers[layer], (32, 9))):
            endchar = ","
            if(i + 1 == len(np.reshape(layers[layer], (32, 9)))):
                endchar = '};\n\n\n'
            print('{9\'b', end="")
            for weight in weight_vec:
                print(weight, end="")
            print("}", end=endchar)
    if 'batch_normal' in layer:
        print(np.shape(layers[layer]))
        gamma = 1
        beta = layers[layer][0]
        mean = layers[layer][1]
        std_dev = layers[layer][2]
        params = mean - beta*std_dev
        export_params = [int(x) for x in params]
        print("{", end="")
        prefix = len(bin(max(export_params))) - 2
        for i, elem in enumerate(export_params):
            endchar = ','
            if (i + 1) == len(export_params):
                endchar = '}'
            if elem < 0:
                elem *= -1
            print(str(prefix) + "'d" + str(elem), end=endchar)
        print("};\n\n\n")

        print("{", end="")
        for index, elem in enumerate(export_params):
            endchar = ', '
            if (index + 1) == len(export_params):
                endchar = '};\n\n\n'
            
            if elem > 0:
                print("2'd1", end=endchar)
            else:
                print("2'd0", end=endchar)
        print("}")
        
    if 'quant_dense_' in layer and np.shape(layers[layer]) == (1, 288, 64):
        
        print(np.shape(layers[layer]))
        print('{', end="")
        for i, weight_vec in enumerate(np.reshape(layers[layer], (288, 64))):
            endchar = ","
            if(i + 1 == len(np.reshape(layers[layer], (288, 64)))):
                endchar = '};\n\n\n'
            print('{64\'b', end="")
            for weight in weight_vec:
                print(weight, end="")
            print("}", end=endchar)
            
    if 'quant_dense_' in layer and np.shape(layers[layer]) == (1, 64,10):
        
        print(np.shape(layers[layer]))
        print('{', end="")
        for i, weight_vec in enumerate(np.reshape(layers[layer], (64, 10))):
            endchar = ","
            if(i + 1 == len(np.reshape(layers[layer], (64, 10)))):
                endchar = '};\n\n\n'
            print('{10\'b', end="")
            for weight in weight_vec:
                print(weight, end="")
            print("}", end=endchar)
```
