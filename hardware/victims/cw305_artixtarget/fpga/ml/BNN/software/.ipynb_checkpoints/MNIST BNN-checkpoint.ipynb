```python
import tensorflow as tf 
import larq as lq
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# Normalize pixel values to be between -1 and 1
train_images, test_images = train_images / 127.5 - 1, test_images / 127.5 - 1
```

```python
# All quantized layers except the first will use the same options
kwargs = dict(input_quantizer="ste_sign",
              kernel_quantizer="ste_sign",
              kernel_constraint="weight_clip")

model = tf.keras.models.Sequential()

# In the first layer we only quantize the weights and not the input
model.add(lq.layers.QuantConv2D(32, (3, 3),
                                kernel_quantizer="ste_sign",
                                kernel_constraint="weight_clip",
                                use_bias=False,
                                input_shape=(28, 28, 1)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.BatchNormalization(scale=False))

model.add(lq.layers.QuantConv2D(64, (3, 3), use_bias=False, **kwargs))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.BatchNormalization(scale=False))

model.add(lq.layers.QuantConv2D(64, (3, 3), use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(tf.keras.layers.Flatten())

model.add(lq.layers.QuantDense(64, use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(lq.layers.QuantDense(10, use_bias=False, **kwargs))
model.add(tf.keras.layers.BatchNormalization(scale=False))
model.add(tf.keras.layers.Activation("softmax"))
```

```python
lq.models.summary(model)
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, batch_size=64, epochs=6)

test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```python
print(f"Test accuracy {test_acc * 100:.2f} %")
```

```python
import numpy as np
preds = model.predict(test_images)

for index, elem in enumerate(preds[0:100]):
    print("Model predicted", test_labels[index], "as", np.argmax(elem), "with probability of", elem.max())
```

```python
layers = {}
for layer in model.layers:
    print(layer.name,"quantized weights:",np.array(layer.weights).shape)
    print(np.digitize(layer.weights, bins=[0]))
    layers[layer.name] = np.digitize(layer.weights, bins=[0])
```

```python
second_weights = np.reshape(layers['quant_conv2d_2'], (64, 64, 9))

print('{', end='')
for weights in second_weights:
    for weight in weights:
        print('{9\'b', end='')
        for w in weight:
            print(w, end='')
        print('},', end='')
print('};')
```

```python
# ADAM Optimizer:
# alpha = 0.001
# beta_1 = 0.9
# beta_2 = 0.999
# epsilon = 1e-7
# BN Parameters:
# gamma = 0 <--- assuming 1 here
# epsilon = 0.001
# beta = 0
# np.shape(layers['batch_normalization'])
gamma = 1
beta = model.layers[2].weights[0]
mean = model.layers[2].weights[1]
std_dev = model.layers[2].weights[2]

params = mean - beta*std_dev

export_params = [int(x) for x in params]
print(export_params)
print("{", end="")
for elem in export_params:
    print("4'd" + str(elem), end=", ")
print("}")

print("{", end="")
for elem in export_params:
    if elem > 0:
        print("1'd1", end=", ")
    else:
        print("1'd0", end=", ")
print("}")
```

```python
gamma = 1
beta = model.layers[5].weights[0]
mean = model.layers[5].weights[1]
std_dev = model.layers[5].weights[2]

print(model.layers[5].name)
params = mean - beta*std_dev

export_params = [int(x) for x in params]
print(export_params)
print("{", end="")
for i, elem in enumerate(export_params):
    enchar = ', '
    if i == len(export_params) - 1:
        enchar = '}'
    if elem < 0:
        elem *= -1
    print("9'd" + str(elem), end=enchar)
print(";")

print("{", end="")
for i, elem in enumerate(export_params):
    endchar = ", "
    if i == len(export_params) - 1:
        endchar = "}"
    if elem > 0:
        print("1'd1", end=endchar)
    else:
        print("1'd0", end=endchar)
print(";")
```
